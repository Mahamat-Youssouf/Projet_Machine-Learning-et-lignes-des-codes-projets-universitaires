```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn import datasets
```

```python
data_vin = datasets.load_wine()
```

```python
print(data_vin)
```

```python
data_vin.keys()
```

```python
print(data_vin['DESCR'])
```

```python
print(data_vin['data'])
```

```python
print(data_vin['target_names'])
```

```python
print(data_vin['target'])
```

```python
df_vin = pd.DataFrame(data_vin['data'], columns = data_vin['feature_names'])
```

```python
df_vin
```

```python
df_vin.info()
```

```python
X = df_vin
y = data_vin['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, 
                                                    random_state = 100)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```

```python
grid_param = {'C': [0.1,1,5,10,15,20,30,40,50,100,1000,10000,100000],
             'gamma': [0.0000001, 0.000001, 0.00001, 0.0001, 0.001, 0.01, 0.1, 1]}
```

```python
grid = GridSearchCV(SVC(), grid_param, verbose = 10)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid_model = grid.best_estimator_
```

```python
grid_predictions = grid_model.predict(X_test)
```

```python
grid_model.score(X_test,y_test)
```

```python
print(confusion_matrix(y_test, grid_predictions))
```

```python
print(classification_report(y_test, grid_predictions))
```
